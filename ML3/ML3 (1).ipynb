### Feature Engineering Label Encoding Label Encoder is a technique that will convert categorical data into numeric data. It takes value in ascending order and converts it into numeric data from 0 to n-1.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cols=health.select_dtypes(include=['object']).columns
print(cols)
le=LabelEncoder()
health[cols]=health[cols].apply(le.fit_transform)
print(health.head(6))
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(health.corr(),annot=True,fmt='.2')
```

### Observation:Variables that are showing some effective correlation are:age, hypertension, heart_disease, ever_married, avg_glucose_level.

```python
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score,classification_report
```

```python
classifier = SelectKBest(score_func=f_classif,k=5)
fits = classifier.fit(health.drop('stroke',axis=1),health['stroke'])
x=pd.DataFrame(fits.scores_)
columns = pd.DataFrame(health.drop('stroke',axis=1).columns)
fscores = pd.concat([columns,x],axis=1)
fscores.columns = ['Attribute','Score']
fscores.sort_values(by='Score',ascending=False)
```

```python
cols=fscores[fscores['Score']>50]['Attribute']
print(cols)
```

```python
train_x,test_x,train_y,test_y=train_test_split(health[cols],health['stroke'])
train_x.shape,test_x.shape,train_y.shape,test_y.shape
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
smote=SMOTE()
train_x,train_y=smote.fit_resample(train_x,train_y)
test_x,test_y=smote.fit_resample(test_x,test_y)
```

```python
print(train_x.shape,train_y.shape,test_x.shape,test_y.shape)
```

### Model Creation

```python
model = LogisticRegression(solver='liblinear', random_state=0)
```

```python
model.fit(train_x,train_y)
predict=model.predict(test_x)
print('Accuracy --> ',accuracy_score(predict,test_y))
print('F1 Score --> ',f1_score(predict,test_y))
print('Classification Report --> \n',classification_report(predict,test_y))
```
