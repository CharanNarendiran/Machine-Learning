### Data Visualization with Python

```python
import pandas as pd 
 
# reading the database
data = pd.read_csv("healthcare.csv")
 
# printing the top 10 rows
display(data.head(8))
```

```python
import matplotlib.pyplot as plt 
# Scatter plot with day against tip 
plt.scatter(data['heart_disease'], data['age'], c=data["avg_glucose_level"]) 
# Adding Title to the Plot plt.title("Scatter Plot") 
# Setting the X and Y Labels plt.xlabel('age') 
plt.ylabel('heart_disease') 
plt.colorbar() 
plt.show() 
```

```python
# Scatter plot with day against tip 
plt.plot(data['age']) 
plt.plot(data['avg_glucose_level']) 
# Adding Title to the Plot plt.title("Scatter Plot") 
# Setting the X and Y Labels plt.xlabel('age') 
plt.ylabel('heart_disease') 
plt.show() 
```

### Bar chart

```python
# Bar chart with day against tip 
plt.bar(data['age'], data['bmi']) 
plt.title("Bar Chart") 
# Setting the X and Y Labels 
plt.xlabel('age') 
plt.ylabel('bmi') 
# Adding the Legends plt.show() 
```

# histogram

```python
 # histogram of total_bills 
plt.hist(data['avg_glucose_level']) 
plt.title("Histogram") 
# Adding the Legends plt.show() 
```
